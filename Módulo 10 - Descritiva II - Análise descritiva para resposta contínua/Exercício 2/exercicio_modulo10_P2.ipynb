# Módulo 10 tarefa 2

Considere a mesma base de dados utilizada na tarefa 1 ```previsao_de_renda.csv```.

1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('previsao_de_renda.csv')
````

````python
df[['posse_de_veiculo', 'posse_de_imovel', 'renda']].corr()
````

Como são poucas variáveis, não é necessário fazer o Clustermap pois, é possível analisar pela matriz de correlação.
Pela Matriz de Correlação é possível concluir que a correlação entre as variáveis é quase nula.

````python
plt.rc('figure', figsize=(10,10))
fig, axes = plt.subplots(2,2)

sns.pointplot(ax = axes[0, 0], x='posse_de_veiculo', y='renda', data=df, dodge=True)
sns.pointplot(ax = axes[0, 1], x='posse_de_imovel', y='renda', data=df, dodge=True)
sns.pointplot(ax = axes[1, 0], x='posse_de_veiculo', y='renda', data=df, hue='posse_de_imovel', dodge=True)
sns.pointplot(ax = axes[1, 1], x='posse_de_imovel', y='renda', data=df, hue='posse_de_veiculo', dodge=True)
````

Pelos gráficos, é possível concluir:

1- Quem possui veículo e quem possui imóvel tem renda maior de quem não possui.<br>
2- Ambas as variáveis são estatisticamente significante, porém, a variável ```pose_de_veiculo``` apresenta um melhor intervalo de confiança.<br>
3- Pela comparação da variável ```pose_de_imóvel``` e ```pose_de_veiculo```, é possível concluir que quem possui veículo tem renda semelhante de quem possui imóvel.
